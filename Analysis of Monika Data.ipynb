Now we will check the variance

```r
not_accurate=remove_outliers_male_med_accuracy[remove_outliers_male_med_accuracy$Medically.accurate==0, ]
med_accurate=remove_outliers_male_med_accuracy[remove_outliers_male_med_accuracy$Medically.accurate==1,]
```

```r
var.test(not_accurate$Males_3_or_more_doses,med_accurate$Males_3_or_more_doses,conf.level = 0.95)
```

Variances are equal

Null Hypothesis: States in which sex education has to be medically accurate do not have higher rates of 3 or more HPV vaccinations than states that do not have medically accurate sex education

Alternative Hypothesis: States in which sex education has to be medically accurate have higher rates of 3 or more HPV vaccinations than states that do not have medically accurate sex education

Welch's t-test for 2 sample tests (one_tailed)

```r
t.test(remove_outliers_male_med_accuracy$Males_3_or_more_doses~remove_outliers_male_med_accuracy$Medically.accurate, var.equal=T, alternative = 'greater', conf.level = 0.95)
```

There does not appear to be a difference in males getting 3 or more HPV shots in states with medically accurate sex education data vs those without
